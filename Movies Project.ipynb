```python
#Follow Along Project with Alex the Analyst on Youtube

# Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)


# Read Data

df = pd.read_csv(r'C:\Users\Foste\Downloads\movies.csv\movies.csv')
```

```python
# Look at Data

df.head()
```

```python
# Check for Missing Data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Drop Missing Data

df = df.dropna(axis=0)

print(df.isnull().sum(),'\n')
```

```python
# Data Types for Columns

df.dtypes
```

```python
# Change Data Types

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Create Column with Correct Year

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df
```

```python
# Order Data

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop Any Duplicates

df.drop_duplicates()
```

```python
# Scatter plot of Budget vs Gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
# Plot Budget vs Gross using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Look for Correlation

df.corr()
```

```python
# Create Correlation Heatmap

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Assign Numbers to Columns with Objects to Allow Correlation

df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
# Create New Correlation Heatmap with All Data

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Unstack Correlation Matrix

correlation_matrix = df_numerized.corr()

corr_pairs = correlation_matrix.unstack()

corr_pairs
```

```python
# Sort Correlation Pairs

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Find Pairs with High Correlation

high_corr = sorted_pairs[(sorted_pairs)>0.5]

high_corr
```
